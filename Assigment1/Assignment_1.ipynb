## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
grouped_data = data[['price', 'grade']].groupby('grade')['price'].apply(list)
plt.figure(figsize=(10, 8))
plt.ticklabel_format(axis="y", style='plain')
plt.xlabel("grade")
plt.ylabel("price")
plt.boxplot(grouped_data, flierprops=dict(markeredgecolor='blue', markerfacecolor='blue', marker='o'));

#Hint: plt.boxplot
```

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

#Your code:
figure, _ = mosaic(data,['bedrooms','floors']);
figure.set_size_inches(10, 8)
figure.get_axes()[0].set_xlabel('bedrooms')
figure.get_axes()[0].set_ylabel('floors')

```

```python
#b)
#Your code:
plt.ticklabel_format(axis="y", style='plain')
plt.xlabel("grade")
plt.ylabel("price")
plt.scatter(data.grade, data.price)
plt.scatter(apartments.grade, apartments.price, color='red', edgecolors='blue', s=100)

```
